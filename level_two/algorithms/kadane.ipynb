```python
# brute force version of finding maximum value of the potential sub-array

def bruteForce(nums: list[int]):
    maxSum = nums[0]

    for i in range(len(nums)): # take ith position as start
        curSum = 0
        for j in range(i, len(nums)): # take jth as last position of sub-array
            curSum += nums[j]
            maxSum = max(maxSum, curSum) # calcualte all sub array values

    return maxSum

# not great because its O(n^2)
```

```python
def kadanes(nums: list[int]):
    maxSum = nums[0]
    curSum = 0

    for n in nums: #iterate through each value in the array
        curSum = max(curSum, 0) # check if cursum is negative, if so reset to 0
        curSum += n # add current element to curSum
        maxSum = max(maxSum, curSum) # if curSum larger than maxSum, it becomes new maxSum
    
    return maxSum
```
