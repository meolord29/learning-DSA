# Graph DFS

Its a type of recursive/backtracking algorithm. very common to see dfs being used in different data structures

```python
def dfs(grid: list[list], row:int, col:int, visit: set):
    rows, cols = len(grid), len(grid[0]) # assume grid is always n*n

    if (
        min(row, col) < 0 or # take values of current x, y - if either negative, you are out of bounds
        row == rows or # if x positon is beyond positive bounds, then you are out of bounds
        col == cols or # if y positon is beyond positive bounds, then you are out of bounds
        (row, col) in visit or # if (x,y) coordinates already exist in the memory, then skip
        grid[row][col] == 1 # since 1 means block, we cannot be at the given (x,y) coordinate
        ):
        return 0
    
    if row == rows - 1 and col == cols - 1: # if we are at the expected location, then at least 1 path exists
        return 1
    
    visit.add((row, col)) # add a note that we have visited this location

    count = 0
    count += dfs(grid, row + 1, col, visit) # moving (y, x+1)
    count += dfs(grid, row - 1, col, visit) # moving (y, x-1)
    count += dfs(grid, row, col + 1, visit) # moving (y+1, x)
    count += dfs(grid, row, col - 1, visit) # moving (y-1, x)

    visit.remove((row, col)) # since its recursive we now want to backtrack, so we want to remove (x,y) from the hash
    return count
    
```
