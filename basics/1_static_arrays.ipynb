```python
class StaticArray:
    def __init__(self, capacity):
        self.capacity = capacity
        self.arr = [0] * self.capacity
        self.length = 0
        
    def insertEnd(self, n):
        if self.length < self.capacity:
            self.arr[self.length] = n
            self.length += 1
        else:
            raise Exception("Not enough capacity in the static array!")
    
    def removeEnd(self):
        if self.length > 0:
            self.arr[self.length - 1] = 0
            self.length -= 1
        else:
            raise Exception("Nothing to remove, the static array is empty!")
    
    def insertMiddle(self, i, n):
        if self.length < self.capacity:
            for index in range(self.length - 1, i-1, -1): # need to shift all elements 1 step to the right
                self.arr[index+1] = self.arr[index]
            
            self.arr[i] = n
            self.length +=1
        else:
            raise Exception("Not enough capacity in the static array!")
```
