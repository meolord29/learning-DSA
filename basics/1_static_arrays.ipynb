```python
class StaticArray:
    def __init__(self, capacity):
        self.capacity = capacity
        self.arr = [0] * self.capacity
        self.length = 0
        
    def insertEnd(self, n):
        if self.length < self.capacity:
            self.arr[self.length] = n
            self.length += 1
        else:
            raise Exception(f"Error - capacity is {self.capacity}, but wanted to add {self.capacity+1}th item")
    
    def removeEnd(self):
        if self.length > 0:
            self.arr[self.length - 1] = 0
            self.length -= 1
        else:
            raise Exception("Nothing to remove, the static array is empty!")
    
    def insertMiddle(self, i, n):
        if self.length < self.capacity:
            for index in range(self.length - 1, i-1, -1): # need to shift all elements 1 step to the right
                self.arr[index+1] = self.arr[index]
            
            self.arr[i] = n
            self.length +=1
        else:
            raise Exception(f"Error - capacity is {self.capacity}, but wanted to add {self.capacity+1}th item")
```

```python
arr = StaticArray(8)
for i in range(8):
    arr.insertEnd(1)
```

```python
arr = StaticArray(8)
for i in range(9):
    arr.insertEnd(1)
```

```python
arr = StaticArray(8)
for i in range(8):
    arr.insertMiddle(1, 1)
```
