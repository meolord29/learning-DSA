```python
class DynamicArray:
    def __init__(self):
        self.capacity = 2
        self.length = 0
        self.arr = [0] * self.capacity
    
    def insertEnd(self, n):
        if self.length == self.capacity:
            self.resize()

        self.arr[self.length] = n
        self.length += 1
    
    def resize(self):
        self.capacity *= 2
        newArr = [0] * self.capacity

        for i in range(self.length):
            newArr[i] = self.arr[i]

        self.arr = newArr

    def popLast(self):
        if self.length > 0:
            self.length -= 1
    
    def get(self, i):
        if i < self.length:
            return self.arr[i]
        else:
            raise Exception(f"Selected Index is larger than array length! {i} > {self.length}")
    
    def insert(self, i, n):
        if i < self.length:
            self.arr[i] = n
            return 0
        else:
            raise Exception(f"Selected Index is larger than array length! {i} > {self.length}")
    
    def print(self):
        for i in range(self.length):
            print(self.arr[i])
```
