# One-Branch
n! = n(n-1)(n-2)...1

```python
def factorial(n):
    # ... missing a backstop
    return n*factorial(n-1)
```

```python
def factorial(n):
    if n==1:
        return 1
    return n*factorial(n-1)
```

Most recursive algorithms can be changed into iteration algorithms, we can have a look at the factorial example

```python
def factorial(n):
    res = 1
    while n>1:
        res *= n
        n-=1
    return res
```
