# Binary Search

Algorithm that takes in an already sorted array of data and then uses tree like logic to look if it should shift to the relative right or left of the given pointer position until value is either found, not found, or closest value to the specified is found.

time complexity of O( log(n) ) due to search space halving at every iteration. Space complexitry is O(1)

```python
def binary_search(arr: list[int], target: int):
    L, R = 0, len(arr) - 1

    while L <= R:

        mid = (L + R) // 2

        # can only be 1 ouf of 3 options at all times, multiple conditions cannot be chacked for via multiple if statements unless you use continue keyword
        if arr[mid] > target:
            R = mid - 1
        elif arr[mid] < target:
            L = mid + 1
        else:
            return mid
    return -1
```

```python
binary_search([1,2,4,6,7,8,12,16,17,25,60], 12)
```
