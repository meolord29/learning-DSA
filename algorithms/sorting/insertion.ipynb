# Insertion sort

Insertion sort is basically looking at an array of elements from left to right. We do not look at the whole array at once but only step by step, increasing the scope of "sorted" array by 1 element at a time. If newly scoped element is less than previous element, we enter a loop that will iteratively move the selected element towards the left until the element on it's left side is either less or equal to the current observed element.

```python
def insertion_sort_first_loop(arr: list[int]):
    for i in range(len(arr)):
        obs_element = arr[i]
        pos_of_comparison = i - 1

        while obs_element > arr[pos_of_comparison] and pos_of_comparison >= 0: # We need to control pos_of_comparision to not go negative since that will mean that all elements on the left have already been observed and compared to obs_element
            arr[pos_of_comparison + 1] = arr[pos_of_comparison] # shifting left neighbor 1 position to the right
            pos_of_comparison -= 1 # shift our focus to the next left element
            
        # once loop is complete, we set obs_element to the right side of the current position of interest. Smallest position of interest can be -1, so actual smallest position of existance in the array is -1+1 = 0
        arr[pos_of_comparison + 1] = obs_element 

    return arr
```

```python
insertion_sort_first_loop([1,7,3,9,2])
```

This sort has worst time complexity of O(n^2) since it has a nested for loop inside the other for loop. Best case O(n) where the list is already sorted. This is also a `stable` sorting algorithm, because it keeps elements will keep their relative position between 2 keys if the key values are identical. this is not always the case across different sorting algorithms.

Also, this sorting algororithm does inplace sorting, which means space complexity is O(1)
