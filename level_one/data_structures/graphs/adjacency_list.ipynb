```python
edges = [["A", "B"], ["B", "C"], ["B", "E"], ["C", "E"]]

adjList = {}
for src, dest in edges:
    if src not in adjList:
        adjList[src] = []
    
    if dest not in adjList:
        adjList[dest] = []
    
    adjList[src].append(dest)
```

```python
# DFS

def dfs(node, target, adjList, visit):
    if node in visit:
        return 0
    if node == target:
        return 1
    
    count = 0

    visit.add(node)

    for neighbor in adjList[node]:
        count += dfs(neighbor, target, adjList, visit)
    visit.remove(node)

    return count
```

```python
from collections import deque

def bfs(node, target, adjList):

    length = 0
    visit = set()
    visit.add(node)
    queue = deque()
    queue.append(node)

    while queue:
        for i in range(len(queue)):
            curr = queue.popleft()
            if curr == target:
                return length
            
            for neighbor in adjList[node]:
                if neighbor not in visit:
                    visit.add(neighbor)
                    queue.append(neighbor)
        
        length += 1
    return length
```
