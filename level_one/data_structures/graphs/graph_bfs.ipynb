# Graph BFS

Its a type of recursive/backtracking algorithm. very common to see dfs being used in different data structures

Used for finding shortest path in the matrix from x0,y0 to x1,y1

```python
from collections import deque
```

```python
def bfs(grid):
    rows, cols = len(grid), len(grid[0])
    visit = set() # remember coordinates bit at
    queue = deque() # track which coordinates left to move from to other 
    queue.append((0,0))
    visit.add((0,0))

    length = 0

    while queue:
        for i in range(len(queue)):
            row, col = queue.popleft()

            if row == rows - 1 and col == cols - 1: # destination achieved
                return length 

            neighbors = [[0, 1], [0, -1], [1, 0], [-1, 0]] # allowed directions

            for dr, dc in neighbors:
                if (
                    min(row + dr, col + dc) < 0 or # same as dfs - check if x or y are out of bounds on negative sides
                    row + dr == rows or # same as dfs - check if x or y are out of bounds on positive sides
                    col + dc == cols or # ^
                    (row + dr, col + dc) in visit or # check if the given x,y was visited before or not
                    grid[row + dr][col + dc] == 1 # check if such position is blocked or not
                    ):

                    queue.append((row + dr, col + dc)) # track of which coords still need to be visited from current coord.
                    visit.add((row + dr, col + dc)) # store a track of which coordinate was visited

            length += 1

```

Time complexity of O(N*M) - this algorithm comes up a lot
