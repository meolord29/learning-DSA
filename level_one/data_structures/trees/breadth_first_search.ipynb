# Breadth Search First

algoriothm where you go level by level instead of all the way down first, then lsowly shift to the right.

```python
from collections import deque
```

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None
```

```python
def bfs(root: TreeNode):

    queue = deque()

    if root: # if root value exists. you store it.
        queue.append(root)

    level = 0 # default level

    while len(queue) > 0: # don't stop until the queue is empty
        print(f"level: {level}")
        for i in range(len(queue)):
            curr = queue.popleft() # get current value in the queue
            if curr.left:
                queue.append(curr.left) # since it's binary tree, you append the node to the queue
            if curr.right:
                queue(curr.right) # since it's binary tree, you append the node to the queue

        # We are now done with current level, we repeat the process with nodes we pushed into the queue
        # once we are donw with those, we will pop it before adding new nodes to the queue. 
        # since we appended them to the queue, they are on the right of the queue,  
        # so leftest one will be the current level, until ith value stops, so we escape the for loop
        level += 1 
```
