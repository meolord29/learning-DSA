# BST

Rule of the game: every left value on the tree must be less than parent value, every right value must be greater than the parent value.

We assume that binary tree does not have duplicated values.

```python
def search(root, value):
    if not root:
        return False
    
    if value > root.value:
        return root(root.right, value)
    elif value < root.value:
        return search(root.left, value)
    else:
        # if value is not greater or less than, then it must be equal
        return True 
```

Why have a tree of array does the same with same time complexity?

Because it is easy to update position of a value in a tree, like link lists have. Where as array, as arrays, all values must be shifted to the right which is basicaly O(N) operation anyway.
