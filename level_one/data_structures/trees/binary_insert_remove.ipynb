```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None
```

```python
def insert(root: TreeNode, v):
    if not root:
        return TreeNode(vallue = v)
    
    if v > root.value:
        root.right = insert(root.right, v)
    elif v < root.value:
        root.left = insert(root.left, v)
    return root
```

```python
def getMinValueNode(root: TreeNode):
    curr = root
    while curr and curr.left:
        curr = curr.left
    return curr
```

```python
def remove(root, value):
    # STEP 1: Base Case
    # If we reach a None node, the value doesn't exist in this path.
    if not root: 
        return None
    


    #### Recursively drill down until find value
    # STEP 2: Search for the node
    # If value is greater than current node, go to the right subtree
    if value > root.value: 
        root.right = remove(root.right, value)
        
    # If value is smaller than current node, go to the left subtree
    elif value < root.value:
        root.left = remove(root.left, value)
    ####



    #### IF YOU FOUND THE VALUE YOU THINK WHAT DO WITH IT
    # STEP 3: Node Found (value == root.value)
    # We must decide how to remove it based on its children
    else:
        # CASE A: Node has NO left child (or is a leaf)
        # We replace this node with its right child.
        # (If right is also None, this correctly returns None, effectively removing the leaf)

        #### IF LEFT IS NONE, THEN RETURN RIGHT, SO YOU MOVE RIGHT VALUE, BUT RIGHT VALUE COULD BE A NONE OR A REAL VALUE - EITHER WAY WINNER MOVE
        if not root.left:  
            return root.right
            
        # CASE B: Node has NO right child
        # We replace this node with its left child.
        #### IF RIGHT IS NONE, THEN RETURN LEFT, SO YOU MOVE LEFT VALUE, BUT LEFT VALUE COULD BE A NONE OR A REAL VALUE - EITHER WAY WINNER MOVE
        elif not root.right:
            return root.left

        # CASE C: Node has TWO children
        # We need to find the "In-Order Successor" (smallest node in the right subtree)
        # to maintain the BST properties.
        #### IF BOTH CHILDREN EXIST, THEN YOU HAVE TO GO TO THE RIGHT MOST LEFT VALUE, GUARANTEED TO BE BEST VALUE DUE TO TREE STRUCTURE
        else:
            # 1. Find the minimum value in the right subtree
            minNode = getMinValueNode(root.right)
            
            # 2. Replace the current node's value with that minimum value
            # (We don't delete the node object, just update the value)
            root.value = minNode.value
            
            # 3. Delete the duplicate node (the minNode) from the right subtree
            # We call remove recursively on the right subtree to remove the value we just moved up.
            root.right = remove(root.right, minNode.value)
            
    # Return the root (which might be the same node, or a replacement child)
    # RETURN ROOT OF THE TREE BACK AFTER ALL OF THE MAGIC HAPPENED (RECURSIVELY)
    return root
```
