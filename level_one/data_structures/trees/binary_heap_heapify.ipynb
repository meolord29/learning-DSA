# Heapify

Special algorithm that takes in an array of values and strucutres it into a heap.

```python
class Heap:
    def __init__(self):
        self.heap = [0] # Initialize heap with a dummy value at index 0 to simplify 1-based indexing
    
    def swap(self, i:int, right = True):
        
        if right:
            right_pos = i * 2 + 1

            tmp = self.heap[i]
            self.heap[i] = self.heap[right_pos]
            self.heap[right_pos] = tmp
            return right_pos

        else:
            left_pos = i * 2

            tmp = self.heap[i]
            self.heap[i] = self.heap[left_pos]
            self.heap[left_pos] = tmp
            return left_pos
        

    def heapify(self, arr):
        arr.append(arr[0]) # push 0th element back to last position, to make math work

        self.heap = arr
        # we can ignore the second half of the elements, 
        # because they won't have children of their own.
        curr = (len(self.heap) - 1) // 2 
        while curr > 0:
            # sink the element downwards
            i = curr
            while 2 * i < len(self.heap):
                left_pos= i * 2
                right_pos = i * 2 + 1
                if (right_pos < len(self.heap) and
                    self.heap[right_pos] < self.heap[left_pos] and
                    self.heap[i] > self.heap[right_pos]):
                    i = self.swap(i, right = True)

                elif self.heap[i] > self.heap[left_pos]:
                    i = self.swap(i, right = False)
                else:
                    break
            
            curr -= 1

    

    
```
