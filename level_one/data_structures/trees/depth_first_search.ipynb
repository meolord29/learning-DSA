# Depth First Search - BST

Recursively go down to the most left element, once you are done, you just move to the right, then go up once you are done, to go donw another branch again.

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None
```

```python
def dfs(root: TreeNode):
    if root is None:
        return ...
    
    dfs(root.left)
    print(root.value) # this action can take place after or before the recursion hits, so it can become preOrder or postOrder actions
    dfs(root.right)
```
