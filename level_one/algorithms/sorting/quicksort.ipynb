# Quick sort

Divide and Conquer with slight changes of what is being compared at what time.

```python
def quicksort(arr):
    if len(arr) < 2: # trivial case
        return arr
    
    low, same, high = [], [], []

    pivot = arr[len(arr) - 1]

    for item in arr:
        if item < pivot:
            low.append(item)
        elif item == pivot:
            same.append(item)
        else:
            high.append(item)

    return quicksort(low) + same + quicksort(high)
```

```python
quicksort([1,8,3,5,2,34,75,95,23,56,12,15])
```

time complexity Nlog(N), but worst case complexity is N^2 as all values can always be on the left/right side so the tree is one sided. But space complexity is O(1) (although greatly depends on the implementation)
