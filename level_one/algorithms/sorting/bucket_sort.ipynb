# Bucket Sort

Basicallly you have lots of numbers, but not too many. you know know the total number of unique values in the array, then you can sort them.

```python
def bucket_sort(arr):

    result = []

    lowest_value = 0
    highest_value = 0

    values = dict()

    for value in arr:

        if value < lowest_value:
            lowest_value = value
        elif value > highest_value:
            highest_value = value


        if value in values:
            values[value] += 1
        else:
            values[value] = 1

    for value in range(lowest_value, highest_value+1):
        if value in values:
            result += [value for _ in range(values[value])]
    
    return result
```

```python
arr = [1,4,6,7,7,7,3,8,16, 24, 55, 55 ,55, 55]
```

```python
bucket_sort(arr)
```
