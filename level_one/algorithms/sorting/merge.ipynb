# Merge Sort

In short, splits array into more splits until its singular units, then sorts them on a micro level, then sorts on a macro level. Beauty is that it works at nlogn complexity. Its our first introduction to divide and conquer algorithm strategy.

```python
# we will start with merge function

def merge(left: list, right:list):
    # merge function goes through both lists, we assume that both left and right arrays are locally sorted, meaning elements in both left and right are sorted with respect to the rest of the elements within the arrays.
    # however there is no guaranteee the left and right are sorted globally meaning that all right elements are equal or greater than elements in the left and are also sorted with respect to each other.

    sorted_array = []
    left_pos, right_pos = 0,0 

    # do not stop the loop until both left and right positions equal to their max array length
    while left_pos != len(left) or right_pos != len(right): 
        # if left pos is equal to max of left array, then obviously just push right array because its already sorted
        if left_pos==len(left): 
            sorted_array.append(right[right_pos])
            right_pos += 1
        # same principle as before but vice versa
        elif right_pos == len(right):
            sorted_array.append(left[left_pos])
            left_pos += 1
        # if neither reached their max limit yet, you compare current right and current left, if current right is bigger then add to sort array
        elif right[right_pos] > left[left_pos]:
            sorted_array.append(right[right_pos])
            right_pos += 1
        # if right position is not bigger so only 1 option left, left position is larger
        else:
            sorted_array.append(left[left_pos])
            left_pos += 1

    return sorted_array
```

```python
def mergeSort(lo, hi, arr):
    if len(arr) == 0: # trivial case 1
        return []
    
    
    if lo == hi:
        return [arr[lo]] # trivial case 2
    
    mid = (lo + hi) // 2 # slice to get mid way but you should slice by // instead of normal division to avoid floating points
    left = mergeSort(lo, mid, arr)
    right = mergeSort(mid+1, hi, arr)

    return merge(left, right)
```
