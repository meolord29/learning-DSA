# Search Range

Imagine we have a custom function that states if our value is too small or to large or correct answer, how would you implement the search?

```python
def checker(n:int):
    k = 25
    if n < k:
        return -1
    elif n > k:
        return 1
    else:
        return 0
```

```python
def binary_search(arr: list[int], method = checker):
    L, R = 0, len(arr) - 1

    while L <= R:
        mid = (L + R) // 2

        if method(arr[mid]) == -1:
            L = mid + 1
        elif method(arr[mid]) == 1:
            R = mid - 1
        else:
            return mid
    
    return -1
```

```python
binary_search([1,2,4,6,7,8,12,16,17,25,60], checker)
```
