# Backtracking

Common problem what we work with binary trees.

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.right = None
        self.left = None
```

```python
# Is there a path from root to leaf node, 
# where all nodes in the path are non zero values

def pathSearch(root: TreeNode):

    # trivial case but also checked at each leaf where leaf is not None
    if not root and root.value == 0: 
        return False
    
    # assuming we are the the leaf node - no children - winner spot
    if not root.left and not root.right:
        return True
    
    # if previous wasn't triggered, then we have to keep going
    if pathSearch(root.left):
        return True
    
    # if previous wasn't triggered, then we have to keep going
    if pathSearch(root.right):
        return True
    
    # if route is ultimately not found, we return False
    return False
    

```
