# Two branch

f(n) = f(n-1) + f(n-2) 

f(0) = 0, f(1)=1

```python
def fib(n):
    if n<=1:
        return n
    
    return fib(n-1) + fib(n-2)
```

```python
%%timeit -n1 -r1

value = fib(1)
print(value)
```

```python
%%timeit -n1 -r1

value = fib(43)
```

```python
class Fib:
    def __init__(self):
        self.mem = dict()

    def save(self, k, v):
        self.mem[k] = v
        

    def run(self, k):
        if k <= 1:
            return k

        elif k in self.mem:
            return self.mem[k]
        else:
            result = self.run(k - 1) + self.run(k - 2)
            self.save(k, result)
            return result
```

```python
%%timeit -n1 -r1
fib_object = Fib()
value = fib_object.run(1)
print(value)
```

```python
%%timeit -n1 -r1
fib = Fib()
value = fib.run(43)
print(value)
```
